```python
def get_user_top_ratings(userID,df):
    u = df[df['userID'] == userID][['songID', 'rating']]
    user_top_ratings=u[u['rating']>4]

    return user_top_ratings
```

```python
def get_similar_users(input_user,df):
    u=get_user_top_ratings(input_user,df)
    
    similar_users=[]
    for song in u['songID']:
        x=df[(df['songID']==song )&(df['rating']>4)]
        for other_user_id in df['userID'].unique():
            if any(x['userID'] == other_user_id):
                similar_users.append(other_user_id)
    return similar_users
        
 
        
```

```python
def recommend_ten_songs(input_user,df):
    u=get_similar_users(0,df)
    filtered_df = df[df['userID'].isin(u)]
    filtered_df = filtered_df[filtered_df['rating'] > 3]
    filtered_df.set_index('songID', inplace=True)
    grouped_df = filtered_df.groupby(level='songID').count()
    rating_list = sorted(grouped_df['rating'],reverse=True)
    ten_songs=[]
    for i in range(10):
        song_id = grouped_df[grouped_df['rating'] == rating_list[i]].index[0]
        ten_songs.append(song_id)
    return ten_songs   
```

```python
recommend_ten_songs(0,df)
```
