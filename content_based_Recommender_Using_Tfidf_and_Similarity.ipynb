```python

df['artist']=df['artist'].str.replace(" ","")
df['text_features'] = df['tag']+" "+df['tag']+" "+df['tag']+" "+df['artist']+" "+df['artist']+" "+df['artist'] +" "+df['artist'] +" "+df['artist'] +" "+df['title']+ " "+df['title']+ " "+df['title']+ " "+df['lyrics']

```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['text_features'])
# random_song=df[df['artist']=='Cairokee']
# random_song.iloc[0]
# text=random_song.iloc[0]['text_features']
# new_text_matrix=vectorizer.transform(['text_features'])
# cosine_similarities = cosine_similarity(new_text_matrix, tfidf_matrix)

# sorted_index=cosine_similarities.argsort()[0][::-1]

# five_most_similar_text=[df.iloc[i]['artist'] for i in sorted_index[1:5]]
# print(five_most_similar_text)
```

```python
def recommendation(IDS):

    texts=df[df['id'].isin(IDS)]['text_features']

    print(texts.shape)
    lst=[]
    for text in texts:

        new_text_matrix=vectorizer.transform([text])
        lst.append(new_text_matrix)
  
    print((lst))
    similarities=[]
    for similarity in lst:
        cosine_similarities = cosine_similarity(similarity, tfidf_matrix)
        similarities.append(cosine_similarities)

     
    most_similar_text={} 
    for array in similarities:
        
        sorted_index=array.argsort()[0][::-1]

        five_most_similar_text=[df.iloc[i]['title'] for i in sorted_index[1:5]]

        most_similar_text[sorted_index[0]]=five_most_similar_text

    return most_similar_text
```
