```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

import io

# Specify the correct file path and delimiter
file_path = r"D:\1\Ali\Machine_Learning\Movie_Recommendation_System\ml-1m\movies.dat"
delimiter = '::'  # The delimiter used in the .dat file

# Read the .dat file using pandas
with open(file_path, 'r', encoding='latin1') as file:
    content = file.read()
    movies = pd.read_csv(io.StringIO(content), delimiter=delimiter, header=None,engine='python')

new_column_names = {
    0: 'movieId',
    1: 'title',
    2: 'genres'
}

movies.rename(columns=new_column_names, inplace=True)

movies

```

```python
movies['year'] = movies['title'].str.extract(r'(\d{4})')
movies
```

```python
import re

def clean_title(title):
    title = re.sub("[^a-zA-Z0-9 ]", "", title)
    return title
movies['clean_title'] = movies['title'].apply(clean_title)
movies['year'] = movies['title'].str.extract(r'(\d{4})')
movies.drop("title",axis=1,inplace=True)
def clean_titleee(title):
    title = re.sub("[^a-zA-Z ]", "", title)
    return title
movies['clean_title'] = movies['clean_title'].apply(clean_titleee)
movies
```

```python
movies['title'] = movies['clean_title']
movies.drop(columns=['clean_title'], inplace=True)
movies
```

```python
# Specify the correct file path and delimiter
file_path = r"D:\1\Ali\Machine_Learning\Movie_Recommendation_System\ml-1m\ratings.dat"
delimiter = '::'  # The delimiter used in the .dat file

# Read the .dat file using pandas
with open(file_path, 'r', encoding='latin1') as file:
    content = file.read()
    ratings = pd.read_csv(io.StringIO(content), delimiter=delimiter, header=None,engine='python')
new_column_names = {
    0: 'userId',
    1: 'movieId',
    2: 'rating',
    3: 'timestamp'
}

ratings.rename(columns=new_column_names, inplace=True)
ratings.drop_duplicates
ratings.isna().sum()
```

```python
ratings
```

```python
# Specify the correct file path and delimiter
file_path = r"D:\1\Ali\Machine_Learning\Movie_Recommendation_System\ml-1m\users.dat"
delimiter = '::'  # The delimiter used in the .dat file

# Read the .dat file using pandas
with open(file_path, 'r', encoding='latin1') as file:
    content = file.read()
    users = pd.read_csv(io.StringIO(content), delimiter=delimiter, header=None,engine='python')
new_column_names = {
    0: 'userId',
    1: 'gender',
    2: 'age',
    3: 'occupation',
    4: 'zip_code'
}

users.rename(columns=new_column_names, inplace=True)
users
```

```python
one_hot_encoded = pd.get_dummies(users['gender'], prefix=None)
users = pd.concat([users, one_hot_encoded], axis=1)
users.drop(columns=['gender'],inplace=True)
users.drop_duplicates()
users.isna().sum()
```

```python
merged_df=pd.merge(pd.merge(ratings, users, on="userId"),movies, on="movieId")
merged_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

merged_df['genres']=merged_df['genres'].str.replace("|","  ")
merged_df['text_features']=merged_df['genres']+" "+merged_df['genres']+" "+merged_df['year']+" "+merged_df['title']+" "+merged_df['title']





def calculate_cosine_similarities(input_id, dataset):
    input_text = dataset[dataset['movieId'] == input_id]['text_features'].iloc[0]

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(dataset['text_features'])

    input_matrix = vectorizer.transform([input_text])

    cosine_similarities = cosine_similarity(input_matrix, tfidf_matrix)

    top_similarities_indices = np.argsort(cosine_similarities, axis=1)[:, -5:][:, ::-1]

    top_similarities_titles = [dataset.iloc[i]['title'] for i in top_similarities_indices[0]]

    return top_similarities_titles

top_similarities = calculate_cosine_similarities(1, merged_df)
print(top_similarities)


```
